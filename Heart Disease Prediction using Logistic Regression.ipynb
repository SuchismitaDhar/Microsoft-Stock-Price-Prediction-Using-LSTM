# Key Risk Factor Analysis

```python
print("\nAnalyzing key risk factors...")
risk_factors = ['age', 'totChol', 'sysBP', 'glucose', 'cigsPerDay']

# Create figure with subplots
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()

for idx, factor in enumerate(risk_factors):
    #Direct separation and plotting
    ax = axes[idx]
    
    # Separate data by CHD status
    data_0 = df[df['TenYearCHD'] == 0][factor].values
    data_1 = df[df['TenYearCHD'] == 1][factor].values
    
    # Create boxplot
    box = ax.boxplot([data_0, data_1], 
                     labels=['No CHD', 'CHD'],
                    patch_artist=True)
    
    # Color the boxes
    box['boxes'][0].set_facecolor('lightblue')
    box['boxes'][1].set_facecolor('lightcoral')
    
    ax.set_title(f'{factor} vs CHD Status')
    ax.set_xlabel('CHD Status')
    ax.set_ylabel(factor)

plt.show()

print("Box plots created successfully!")
```

# Building Machine Learning  Model

```python
#Looking into our target variable
df["TenYearCHD"].value_counts()
```

There is great imbalance in the nos. "Not having Heart Disease" and "Having Heart Disease"

### Logistic Regression Model

```python
X=df_cont
Y=df["TenYearCHD"]

#Splitting Data into Train(70%) and Test(30%)
from sklearn.model_selection import train_test_split
X_train,X_test, y_train, y_test=train_test_split(X,Y,test_size=0.3, random_state= 100)

#Scaling the Train Features fro Optimal Model Performance
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train_sc=sc.fit_transform(X_train)
print ("Scaled Training Input Data:", X_train_sc)
 
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X_train_sc,y_train)
```

## Display Model Coefficients

```python
feature_imp=pd.DataFrame({"Features":X.columns,"Coefficient": lr.coef_[0]})
feature_imp
```

### Testing the Model

```python
X_test_sc=sc.transform(X_test)
y_lr_train_pred=lr.predict(X_train_sc)
y_lr_test_pred=lr.predict(X_test_sc)
y_pred_prob=lr.predict_proba(X_test_sc)
print(y_pred_prob)
y_test_arr=np.array(y_test).ravel()
print("\n")
y_test_arr
```

# Model Evaluation

### 1-Precision Score
### 2- Accuracy Score
### 3- Recall Score
### 4- Confusion Matrix

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import (precision_score, recall_score, confusion_matrix)

lr_train_prec=precision_score(y_train, y_lr_train_pred)
lr_train_acc=accuracy_score(y_train, y_lr_train_pred)
lr_train_f1=f1_score(y_train, y_lr_train_pred)
lr_train_rec=recall_score(y_train, y_lr_train_pred)

lr_test_prec=precision_score(y_test_arr, y_lr_test_pred)
lr_test_acc=accuracy_score(y_test_arr, y_lr_test_pred)
lr_test_f1=f1_score(y_test_arr, y_lr_test_pred)
lr_test_rec=recall_score(y_test_arr, y_lr_test_pred)

cm=confusion_matrix(y_test_arr, y_lr_test_pred)

print  ("LR Precision (Train):", lr_train_prec)      
print ("LR Accuracy Score (Train):", lr_train_acc)
print ("LR F1 Score (Train):", lr_train_f1)
print ("LR Recall Score (Train):", lr_train_rec)
print ("\n")
print ("Confusion Matrix:", cm)
print("\n")
print  ("LR Precision (Test):", lr_test_prec)
print ("LR Accuracy (Test):", lr_test_acc)
print ("LR F1 Score (Test):", lr_test_f1)
print ("LR Recall Score (Test):", lr_test_rec)
```

### Visualize Confusion Matrix

```python
plt.figure(figsize=(6,4))
sns.heatmap(cm,annot=True, xticklabels=["NO CHD", "CHD"], yticklabels=["NO CHD", "CHD"])
plt.xlabel("Predicted Label")
plt.ylabel("Actual Label")
plt.title("Confusion Matrix")
plt.show()
```
